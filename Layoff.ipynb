```python
import requests, re
from bs4 import BeautifulSoup
from csv import writer, reader
from time import sleep


def scraper():
    
#     my_alphabet = ['a','b','c','d', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 
#                    'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 
#                    'v', 'w', 'x', 'y', 'z']

    my_alphabet = ['a']
    
    for letter in my_alphabet:
        file_name = letter + '-companies.csv'
        url = 'https://www.thelayoff.com/' + letter + '.php'
        sleep(10)

        response = requests.get(url)

        soup = BeautifulSoup(response.text, 'html.parser')

        page_counter = 0

        with open(file_name, 'w', encoding='utf-8', newline = '') as csv_file:   # opens new file for each letter
            csv_writer = writer(csv_file)
            csv_writer.writerow(['company_name', 'link', 'ticker'])

            while soup.find(class_='next') is not None:
                page_counter += 1
                response = requests.get(url + '?page=' + str(page_counter)) 
                soup = BeautifulSoup(response.text, 'html.parser')
                companies = soup.find(class_ = 'abc-list')('li')

                for item in companies:
                    company_name = (item.find('a').get_text())
                    link = 'https://www.thelayoff.com/' + item.find('a')['href']

                    #ticker regex logic
                    my_string = item.get_text()
                    my_regex = re.compile(r'Ticker: [A-Z][A-Z](.*)')
                    match = my_regex.search(my_string)
                    if match:
                        temp_ticker = (match.group()[8:-1])   # need to convert to string object to check if NASDAQ, NYSE
                        if 'NASDAQ' in temp_ticker:
                            ticker = (match.group()[15:-1])
                        elif 'NYSE' in temp_ticker:
                            ticker = (match.group()[13:-1])
                        else:
                            ticker = (match.group()[8:-1])
                    else:
                        ticker = ''

                    csv_writer.writerow([company_name, link, ticker])
```

```python
scraper()
```
