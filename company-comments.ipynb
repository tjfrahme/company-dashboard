```python
# Open new excel file
# For loop through every company
# get link from df
# scrape page
# put into excel

import requests
import pandas as pd
from bs4 import BeautifulSoup
import csv
from csv import writer, reader
from time import sleep


def scrape_comments():
    
#     my_alphabet = ['a','b','c','d', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 
#                'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 
#                'v', 'w', 'x', 'y', 'z']
    
    
    my_alphabet = ['v', 'w', 'x', 'y', 'z']

    for letter in my_alphabet:
        source_file = letter + '-companies.csv'
        df = pd.read_csv(source_file)
        df.dropna(inplace = True)
        
        file_name = letter + '-comments.csv'
        sleep(2)
    
        with open(file_name, 'w', encoding='utf-8', newline = '') as file:    #newline removes spaces when writing to csv
            csv_object = csv.writer(file)
            csv_object.writerow(['company', 'ticker', 'title', 'body', 'date', 'replies'])  #Header

            for index, row in df.iterrows():                          # loops through each company 
                response = requests.get(row['link'])
                soup = BeautifulSoup(response.text, 'html.parser')
                articles = soup.find_all('article')

                for article in articles:                  #Loops through the page and copies top comments

                    company = row['company_name']

                    ticker = row['ticker']

                    title = article.find('a').get_text()

                    try:
                        body = article.find(class_='post-body').get_text().strip()
                    except:
                        body = 'no body text'

                    date = article.find('time')['datetime'][:10]

                    try:
                        replies = (article.find(class_='replies').get_text())
                        if replies == 'no replies yet':
                            replies = 0
                        else:
                            replies = article.find(class_='replies').get_text().split()[0]
                    except:
                        replies = 0

                    csv_object.writerow([company, ticker, title, body, date, replies])
```

```python
scrape_comments()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import csv
from csv import writer, reader

my_alphabet = ['b','c','d', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 
               'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 
               'v', 'w', 'x', 'y', 'z']

def combine():

    df1 = pd.read_csv('a-comments.csv')

    for letter in my_alphabet:
        source_file = letter + '-comments.csv'
        df2 = pd.read_csv(source_file)
        df1 = pd.concat([df1, df2])


    with open('all-comments.csv', 'w', encoding='utf-8', newline = '') as file:    #newline removes spaces when writing to csv
        csv_object = csv.writer(file)
        csv_object.writerow(['company', 'ticker', 'title', 'body', 'date', 'replies'])  #Header

        for index, row in df1.iterrows():
            company = row['company']
            ticker = row['ticker']
            title = row['title']
            body = row['body']
            date = row['date']
            replies = row['replies']
            
            csv_object.writerow([company, ticker, title, body, date, replies])
            
```

```python
combine()
```
