```python
from csv import writer, reader

response = requests.get('https://www.thelayoff.com/general-motors')

soup = BeautifulSoup(response.text, 'html.parser')

articles = soup.find_all('article')

with open('gm.csv', 'w', newline = '') as file:    #newline removes spaces when writing to csv
    csv_object = csv.writer(file)
    csv_object.writerow(['title', 'body', 'date', 'replies'])  #Header
    
    for article in articles:                  #Loops through the page
        title = article.find('a').get_text()
        
        try:
            body = article.find(class_='post-body').get_text().strip()
        except:
            body = 'no body text'
        
        date = article.find('time')['datetime'][:10]
        
        try:
            replies = (article.find(class_='replies').get_text())
            if replies == 'no replies yet':
                replies = 0
            else:
                replies = article.find(class_='replies').get_text().split()[0]
        except:
            replies = 0
    
        csv_object.writerow([title, body, date, replies])
```
